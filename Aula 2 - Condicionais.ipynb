## 1. Uma breve contextualiza√ß√£o sobre condicionais
Nesta aula, vamos aprender um dos conceitos fundamentais da programa√ß√£o, que nos permite fazer programas inteligentes, que tomam decis√µes: condicionais.

````python
x = int(input("Escreva um n√∫mero"))

if x % 2 == 0:
  print("Esse n√∫mero √© par")
elif x % 2 == 1:
  print("Esse n√∫mero √© √≠mpar")
````

O programa acima pede um n√∫mero para o usu√°rio, o converte em um n√∫mero inteiro e devolve para o usu√°rio a informa√ß√£o se tal n√∫mero √© par ou √≠mpar.
√â v√°lido mencionar o operador `%` (m√≥dulo), que retorna o resto de uma divis√£o inteira.



No caso de uma divis√£o por 2, o resto ser√° sempre 0 se o n√∫mero for par (ou seja, divis√≠vel por 2) e ser√° 1 se o n√∫mero for √≠mpar (ou seja, n√£o divis√≠vel por dois).



Com isso esclarecido, vamos destrinchar como o bloco condicional em si funciona!

`if` significa 'se', e essencialmente apenas roda se a condi√ß√£o que vem depois do "if" for verdadeira. 

Tente rodar a c√©dula abaixo:

````python
x = 5

if x > 10: # se x for maior que 10
  print('x maior que 10')
````

Perceba que nada acontece, n√£o h√° uma sa√≠da. Agora tente rodar a c√©dula abaixo:

````python
y = 15

if y > 10: # se y for maior que 10
  print('y maior que 10')
````

Agora a informa√ß√£o apareceu, porque a condi√ß√£o de "x maior que 10" foi verdadeira. N√≥s podemos ver isso se pedirmos ao Python para imprimir as condi√ß√µes. A primeira condi√ß√£o retorna False (falso) e a segunda retorna True (verdadeiro).

Esses valores True e False s√£o chamados de valores Booleanos.

````python
x = 5
y = 15
print("O valor de x>10 √©", x>10)
print("O valor de y>10 √©", y>10)
````

Ent√£o, voltando para a checagem de par ou √≠mpar, o python apenas retornar√° o que estiver dentro da condi√ß√£o if se o valor booleano for True.

````python
x = 9 # experimente mudar o valor de x

if x % 2 == 0: # se isso for verdade, o programa imprime que o n√∫mero √© par
  print("Esse n√∫mero √© par")
elif x % 2 == 1: # se isso for verdade, o programa imprime que o n√∫mero √© √≠mpar
  print("Esse n√∫mero √© √≠mpar")
````

### 1.1. Operadores de compara√ß√£o
Operadores de Compara√ß√£o:

`==` (√â igual a?) - Ex: `status_pagamento == 'Realizado'`

`!=` (√â diferente de?) - Ex: `risco != 'Baixo'`

`>` (Maior que) - Ex: `valor_empenho > 50000`

`>=` (Maior ou igual a) - Ex: `valor_empenho >= 50000`

`<` (Menor que) - Ex: `valor_empenho < 50000`

`<=` (Menor ou igual a) - Ex: `valor_empenho <= 50000`

### 1.2. Operadores l√≥gicos
E se uma decis√£o depender de m√∫ltiplas condi√ß√µes? Por exemplo, "Aprovar uma opera√ß√£o SE o risco for baixo E o valor estiver dentro do limite". Para isso, usamos os operadores l√≥gicos:

`and`: Verdadeiro somente se AMBAS as condi√ß√µes forem True.

`or`: Verdadeiro se PELO MENOS UMA das condi√ß√µes for True.

`not`: Inverte o valor Booleano (True vira False e vice-versa).

Os conceitos s√£o os mesmos das tabelas verdade no contexto da mat√©ria de Racioc√≠cio L√≥gico e Matem√°tico.



````python
# Cen√°rio de an√°lise de cr√©dito
score_cliente = 750
possui_inadimplencia = False

# Usando 'and'
aprovacao_automatica = score_cliente > 700 and possui_inadimplencia == False
print(f"Aprova√ß√£o autom√°tica? {aprovacao_automatica}")

# Usando 'or'
alerta_manual = score_cliente < 400 or possui_inadimplencia == True
print(f"Requer an√°lise manual? {alerta_manual}")

# Usando 'not'
# 'not possui_inadimplencia' √© o mesmo que 'possui_inadimplencia == False'
cliente_adimplente = not possui_inadimplencia
print(f"Cliente est√° adimplente? {cliente_adimplente}")
````

## 1.4. Al√©m do if (elif e else)

Por enquanto, n√≥s falamos primariamente sobre o if: se algo for verdadeiro, retorne alguma coisa. 

O python tamb√©m possui duas outras estruturas condicionais: `elif` e `else`. 

Tente mudar o valor da vari√°vel `idade` abaixo:

````python
idade = 18 # tente mudar o valor da idade aqui

if idade < 2:
  print('Voc√™ √© um beb√™.')
elif idade < 12:
  print('Voc√™ √© uma crian√ßa.')
elif idade < 17:
  print('Voc√™ √© um adolescente.')
elif idade < 65:
  print('Voc√™ √© um adulto.')
else: 
  print('Voc√™ √© um idoso.')
````

Perceba que voc√™ nunca retorna mais de uma sa√≠da, mesmo o programa dando 5 op√ß√µes. Por qu√™?

Quando o python roda um `if`, ele nunca vai rodar um `elif` ou `else`. Apenas quando o if n√£o rodar, o python checa as outras op√ß√µes. 




````python
idade = 10 # tente mudar o valor da idade aqui

if idade < 2:
  print('Voc√™ √© um beb√™.')
elif idade < 12:
  print('Voc√™ √© uma crian√ßa.')
elif idade < 17:
  print('Voc√™ √© um adolescente.')
elif idade < 65:
  print('Voc√™ √© um adulto.')
else: 
  print('Voc√™ √© um idoso.')
````

Da mesma maneira, quando um dos elifs for verdade o python ignora tudo que vier depois. 


````python
idade = 70 # tente mudar o valor da idade aqui

if idade < 2:
  print('Voc√™ √© um beb√™.')
elif idade < 12:
  print('Voc√™ √© uma crian√ßa.')
elif idade < 17:
  print('Voc√™ √© um adolescente.')
elif idade < 65:
  print('Voc√™ √© um adulto.')
else: 
  print('Voc√™ √© um idoso.')
````

Por fim, o `else` apenas roda quando todos os `ifs` e `elifs` anteriores n√£o rodarem.
