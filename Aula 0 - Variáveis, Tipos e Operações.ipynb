### 1.2. Vari√°veis
Nesta aula, vamos aprender o conceito mais fundamental da programa√ß√£o: vari√°veis.

Pense em uma vari√°vel como uma "caixa" na mem√≥ria do computador. N√≥s damos um nome (uma etiqueta) para essa caixa e guardamos uma informa√ß√£o dentro dela. Isso nos permite usar e manipular essa informa√ß√£o facilmente mais tarde.

### **1.3. Como nomear vari√°veis**

Existem algumas regras e conven√ß√µes para criar vari√°veis no Python. Pense nisso como a gram√°tica e o estilo de escrita de um idioma: as **regras** s√£o obrigat√≥rias para que o "computador" entenda o que voc√™ quer dizer, enquanto as **conven√ß√µes** s√£o as boas pr√°ticas que tornam seu c√≥digo leg√≠vel para outros seres humanos (e para voc√™ mesmo no futuro).

#### **Regras Obrigat√≥rias (O que causa erro se n√£o for seguido)**

Se voc√™ quebrar uma destas regras, seu c√≥digo n√£o ir√° funcionar e o Python apresentar√° um erro de sintaxe (`SyntaxError`).

  * **1. Deve come√ßar com uma letra ou um underscore (`_`).**

      * Nomes de vari√°veis **nunca** podem come√ßar com um n√∫mero.
      * **Certo:** `ano2025`, `_taxa_selic`
      * **Errado:** `3num = 3`

  * **2. Pode conter apenas letras, n√∫meros e underscores.**

      * Caracteres especiais como h√≠fen (`-`), ponto de exclama√ß√£o (`!`), espa√ßo (`     `), ou qualquer outro s√≠mbolo n√£o s√£o permitidos.
      * **Certo:** `valor_da_divida`, `relatorio_final`
      * **Errado:** `nome-da-pessoa = "Jo√£o"`, `numero! = 6`

  * **3. N√£o pode ser uma palavra reservada do Python.**

      * O Python tem um conjunto de palavras que ele usa para sua pr√≥pria sintaxe (como `def`, `if`, `for`, `while`, `return`, etc.). Voc√™ n√£o pode us√°-las como nomes de vari√°veis.
      * **Certo:** `definicao = "Fun√ß√£o"`
      * **Errado:** `def = "Fun√ß√£o"`

-----

#### **Boas Pr√°ticas e Conven√ß√µes (O que torna o c√≥digo leg√≠vel e "Pyth√¥nico")**

Quebrar estas conven√ß√µes geralmente n√£o causa um erro imediato, mas pode levar a bugs dif√≠ceis de encontrar e torna o c√≥digo confuso.

  * **1. Use nomes descritivos e em min√∫sculas, com palavras separadas por underscore (`_`).**

      * Este padr√£o √© chamado de **snake\_case** e √© a conven√ß√£o oficial para vari√°veis em Python (definida na PEP 8). Um bom nome de vari√°vel deve descrever claramente o que ela armazena.
      * **Bom:** `valor_total_arrecadado`, `taxa_juros_anual`, `nome_do_credor`
      * **Ruim:** `v`, `x`, `vta` (muito curtos), `valorTotalArrecadado` (padr√£o de outra linguagem, chamado camelCase)

  * **2. N√£o use nomes de fun√ß√µes j√° existentes no Python.**

      * Tecnicamente, o Python permite que voc√™ fa√ßa `print = 1`, mas isso √© uma p√©ssima ideia. Ao fazer isso, voc√™ **sobrescreve** a fun√ß√£o original `print()`, e ela deixa de funcionar como esperado.
      * **C√≥digo que quebra:**
        ```python
        # N√ÉO FA√áA ISSO!
        print = "Relat√≥rio Final" 

        # A linha abaixo agora vai gerar um erro, pois 'print' n√£o √© mais uma fun√ß√£o.
        print("Este texto n√£o ser√° exibido") 
        # TypeError: 'str' object is not callable
        ```

  * **3. Nomes s√£o sens√≠veis a mai√∫sculas e min√∫sculas (*case-sensitive*).**

      * As vari√°veis `Capital`, `capital` e `CAPITAL` s√£o tr√™s vari√°veis completamente diferentes para o Python. A conven√ß√£o √© usar apenas letras min√∫sculas para evitar confus√£o.
      * **Exemplo:**
        ```python
        capital = 10000
        Capital = "Bras√≠lia"

        # A linha abaixo causaria um erro, pois a vari√°vel 'Capital' cont√©m um texto.
        # juros = Capital * 0.1 
        ```

  * **4. Nomear constante com todas as letras mai√∫sculas**

      * Enquanto o Python n√£o possui um tipo `const` como outras linguagens de programa√ß√£o, √© uma conven√ß√£o que os programadores usam de que valores que n√£o devem mudar ao longo do c√≥digo s√£o escritos com "ALL CAPS".
      * **Exemplo:**
        ```python
        PI = 3.1415 # note que PI est√° em ALL CAPS
        raio = 2
        area = PI * raio ** 2
        ```

### 1.4. Tipos de dados
O tipo de dados exemplificado acima √© o de n√∫mero inteiro, mas existem muitos outros tipos. Agora vamos ver os tr√™s mais b√°sicos:

Inteiro (`int`): N√∫meros inteiros, sem parte decimal. Perfeito para contagens, como anos ou n√∫mero de meses.

Ponto Flutuante ou Decimal (`float`): N√∫meros que possuem uma parte decimal. Usamos o ponto (`.`) para separar a parte inteira da decimal, diferentemente do portugu√™s (em que usamos uma v√≠rgula `,` como separador decimal. Essencial para valores monet√°rios, taxas, etc.

Texto (`string`): Qualquer sequ√™ncia de caracteres (letras, n√∫meros, s√≠mbolos). O texto deve sempre estar entre aspas (simples ' ou duplas ").

Podemos usar a fun√ß√£o `type()` para perguntar ao Python qual √© o tipo de uma vari√°vel.

````python
# Vari√°veis com diferentes tipos de dados
ano = 2025                          # int (inteiro)
taxa_selic = 10.5                   # float (decimal)
nome_titulo = "Tesouro Selic 2029"    # string (texto)

# Vamos verificar os tipos de cada uma
print("O tipo da vari√°vel 'ano' √©:", type(ano))
print("O tipo da vari√°vel 'taxa_selic' √©:", type(taxa_selic))
print("O tipo da vari√°vel 'nome_titulo' √©:", type(nome_titulo))
````

Divis√£o: `/`
